```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')

soup
```

```python
#Scrape the NASA Mars News Site and collect the latest News Title and 
#Paragraph Text. Assign the text to variables that you can reference later

result = soup.find('div',class_='box floating_text_area left ms-layer')

result


news_title = result.find('h1', class_='media_feature_title').text
news_p = result.find('div', class_='description').text
    
print('-----------------')
print(news_title)
print(news_p)

```

```python
#Mars Hemisphere pics

url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
browser.visit(url)
```

```python
all_hems= []

for i in range(4):
    
    browser.find_by_tag('h3')[i].click()
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    title = soup.find('h2', class_='title').text


    img_url = soup.find('a', text='Sample')['href']


    hem_urls={
        "title": title,
        "img_url": img_url
    }
    
    all_hems.append(hem_urls)
    
    browser.back()
    
print(all_hems)
    

    
    
    
    
```
